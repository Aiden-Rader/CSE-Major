This this the Hardware Teleportation Experiment code, this is not the main code for the project, however it is used to test the fidelity of the teleportation protocol on a real quantum computer using IBM's Qiskit SDK.

```python
# Import necessary modules for hardware runs
import pandas as pd
from teleportation_module import (
	prep_zero, prep_one, prep_plus, prep_minus,
	run_on_hardware, get_ideal_state, fidelity_from_counts
)

print("Imports complete!")
```

```python
# Define Inputs
inputs = {
	"|0>": prep_zero,
	"|1>": prep_one,
	"|+>": prep_plus,
	"|->": prep_minus
}

hardware_rows = []
for label, prep in inputs.items():
	print(f"Running on hardware for input: {label} ...")

	counts = run_on_hardware(prep)           # execute on real IBM chip
	ideal = get_ideal_state(prep)            # ideal target state
	F = fidelity_from_counts(counts, ideal)  # approximate fidelity from measurements

	hardware_rows.append({
		"Input State": label,
		"Hardware Fidelity": F,
		"Raw Counts": counts
	})

pd.DataFrame(hardware_rows)
```
